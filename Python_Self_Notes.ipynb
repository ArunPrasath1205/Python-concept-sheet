####Casting in Python
* If you want to specify the data type of a variable, this can be done with casting.
* Python is an object-orientated language, and as such it uses classes to define data types, including its primitive types.

Casting in python is therefore done using constructor functions:

* **int()** - constructs an integer number from an integer literal, a float literal (by removing all decimals), or a string literal (providing the string represents a whole number)
* **float()** - constructs a float number from an integer literal, a float literal or a string literal (providing the string represents a float or an integer)
* **str()** - constructs a string from a wide variety of data types, including strings, integer literals and float literals

```python
x = str(3)    
y = int(3)    
z = float(3)
print(x,y,z)
```

###Global Variables
* Global variable that has been created outside the function
* That can be used outside/inside the functions

Local vs Global Variables
* If you create a variable with the same name inside a function, this variable will be local, and can only be used inside the function. 
* The global variable with the same name will remain as it was, global and with the original value.

```python
x = "awesome"

def myfunc():
  x = "fantastic" #Local variable
  print("Python is " + x)

myfunc()

print("Python is " + x) # here x is Global variable
```

* Also, use the global keyword if you want to change a global variable inside a function.

```python
def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

####List comprehension

image.png

Expression

* The expression is the current item in the iteration, but it is also the outcome, which you can manipulate before it ends up like a list item in the new list:

* The expression can also contain conditions, not like a filter, but as a way to manipulate the outcome:

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

newlist = [x for x in fruits if "a" in x]

print(newlist)
```

```python
# The expression can also contain conditions, not like a filter, but as a way to manipulate the outcome:

newlist = [x if x != "banana" else "orange" for x in fruits]
newlist
#The expression in the example above says:
#"Return the item if it is not banana, if it is banana return orange".
```

Sorting vs case sensitivity
* By default sorting is case-sensitive 
* But, we can alter the built-in-function as case-insensitive by using **sort(key = str.lower)**  

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort(key = str.lower)
print(thislist)
```
